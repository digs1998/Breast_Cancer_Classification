```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV
import pingouin as pg
```

```python
data = pd.read_csv('/Users/Lenovo/Desktop/Education/Winter Term/QBS-177/data.csv')
data.head(5)
```

```python
data.describe()
```

```python
data.info()
```

```python
breast_data = data.drop('Unnamed: 32', axis=1)
breast_data.isnull().sum()
```

```python
breast_data['diagnosis'] = np.where(breast_data['diagnosis'] == 'M', 1, 0)
```

```python
import seaborn as sns

dummy_breast = breast_data.drop('id', axis=1)
p_cor = dummy_breast.pcorr()

xticklabels = list(p_cor.columns.values)
yticklabels = list(p_cor.columns.values)

xticklabels.insert(2, '')
yticklabels.insert(2, '')

# Create the heatmap
plt.figure(figsize=(6,5))
sns.heatmap(p_cor, cmap='coolwarm', center=0, xticklabels=xticklabels, yticklabels=yticklabels)

# Show the plot
plt.show()
```

```python
breast_data['diagnosis'].value_counts().plot(kind='bar')
```

```python
from sklearn.preprocessing import MinMaxScaler

mx = MinMaxScaler()

x = dummy_breast.iloc[:,1:].values
y = dummy_breast['diagnosis'].values

x_transformed = mx.fit_transform(x)

xtrain, xtest, ytrain, ytest = train_test_split(x_transformed, y, test_size=0.2, random_state=45)
```

```python
xtrain.shape
```

```python
#Using LassoCV

lasso_model = LassoCV(cv=5, random_state=40)
lasso_model.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import f1_score

coef = pd.Series(lasso_model.coef_, index=breast_data.columns[2:])

best_cols = coef[coef != 0].index.tolist()

ypred = lasso_model.predict(xtest)

# calculate the accuracy of the predictions
accuracy = f1_score(ytest, ypred.round())

# print the selected columns and the accuracy of the predictions
print("Selected columns: " + ", ".join(best_cols))
print("Accuracy:", round(accuracy, 4))
```

```python
imp_coef = coef.sort_values()
plt.rcParams['figure.figsize'] = (5.0, 6)
imp_coef.plot(kind = "barh")
plt.title("Features selected by Lasso")
```

```python
best_cols.append('diagnosis')
```
